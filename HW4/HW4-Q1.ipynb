```python
import numpy as np
import pandas as pd
```

```python
cancer_df = pd.read_csv('Cancer.csv')
cancer_df[0::20]
```

```python
# import splitting function
from sklearn.model_selection import train_test_split

# Retrieve all columns minus the Malignant Cancer column
feature_names = (set(cancer_df))
feature_names.remove('Malignant_Cancer')

X = cancer_df[list(feature_names)]
y = cancer_df['Malignant_Cancer']

# Create the training and testing sets
(X_train, X_test, y_train, y_test) = train_test_split(X, y, test_size=0.35, random_state=3)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Create and test the accuracy of Decision Tree Classifier
my_DecisionTree = DecisionTreeClassifier(random_state=3)
my_DecisionTree.fit(X_train, y_train)

y_predict_dt  = my_DecisionTree.predict(X_test)
score_dt  = accuracy_score(y_test, y_predict_dt)

print("Decision Tree: ", score_dt)
```
